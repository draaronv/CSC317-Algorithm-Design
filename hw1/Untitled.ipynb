```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
n=np.linspace(2,100,100)
bubble_sort=8*n**2
heap_sort=64*n*(np.log2(n))
```

```python
fig, (axs1,axs2,axs3) = plt.subplots(1,3)
fig.tight_layout()
axs1.plot(n, bubble_sort)
axs2.plot(n,heap_sort)
idx = np.argwhere(np.diff(np.sign(bubble_sort - heap_sort))).flatten()
axs3.plot(bubble_sort[idx], heap_sort[idx], 'ro')
axs1.set_title('Bubble Sort',fontsize=11)
axs2.set_title('Heap Sort',fontsize=11)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(2, 1000)
f = 8*x**2
g = 512*x*(np.log2(x))
plt.plot(x, f, '-',label="Bubble Sort")
plt.plot(x, g, '-', label="Heap Sort")
idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
plt.plot(x[idx], f[idx],'ro',label= 'point')
plt.xlabel('n', fontsize=18)
plt.ylabel('Number of Steps', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
print(idx)
```
